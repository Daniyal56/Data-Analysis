```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('D:\\Dataset\\googleplaystore.csv',nrows=1000)
```

```python
df.head()
```

```python
x = df['Rating']
y = df['Reviews']
z = df['App']
```

```python
plt.figure(figsize=(16,9))
class_index = np.arange(len(z))
plt.scatter(x,y,label=['Rating', 'Review'],marker='*',c ='r',s=200,alpha=0.5,linewidths=2,edgecolors='k')
plt.scatter(x,df['Installs'],label=['Rating', 'Review'],marker='*',c ='r',s=200,alpha=0.5,linewidths=2,edgecolors='k')
plt.title('Google Play Store Apps Analysis', fontsize=20)
plt.xlabel('Rating',fontsize=15)
plt.ylabel('Review & Install',fontsize=15)
plt.xticks(class_index,z,rotation=50)
plt.show()
```

```python
short_data = pd.read_csv('D:\\Dataset\\googleplaystore.csv',nrows=3000)
```

```python
group = short_data.groupby('Category')
```

```python
cat = group.groups
len(cat)
```

```python
for x,y in group:
    print(x)
    print(y)
```

```python
plt.figure(figsize=(16,9))
class_index = np.arange(len(cat))
width=0.2
plt.scatter(short_data['Rating'],short_data['Category'],marker='*',s=200,c='r',alpha=0.5,linewidths=1,edgecolors='k',label=['Rating','Category'])
plt.yticks(class_index,cat,rotation=50)
plt.title('GooGle Play Store Analysis by using Scatter',fontsize=20)
plt.xlabel('Ratings',fontsize=15)
plt.ylabel('Category',fontsize=15)
plt.legend()
plt.show()
```
