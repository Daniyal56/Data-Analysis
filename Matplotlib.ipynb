# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml = np.random.randint(10,45,(100))
py = np.random.randint(9,38,(100))
```

```python
display(ml,py)
```

```python
plt.figure(figsize=(16,9))
bins = [10,15,20,25,30,35,40,45]
plt.hist(ml,bins,rwidth=0.8,histtype='step',color='k',align='right')
plt.title('Machine Learning')
plt.xlabel('Student Age')
plt.ylabel('No. of Students (AGE)')
plt.show()
```

```python
plt.figure(figsize=(16,9))
bins = [10,15,20,25,30,35,40,45]
plt.hist(ml,bins,rwidth=0.8,histtype='bar', color='k',align='right',orientation='horizontal',label='ML')
#plt.hist(py,bins,rwidth=0.8,histtype='bar', color='o',align='right',orientation='horizontal',label='PY')
plt.title('Machine Learning')
plt.xlabel('Student Age')
plt.ylabel('No. of Students (AGE)')
plt.legend(loc=4)
plt.show()
```

```python
plt.figure(figsize=(16,9))
bins = [10,15,20,25,30,35,40,45]
plt.hist(ml,bins,rwidth=0.8,histtype='bar', color='r',orientation='vertical',label='PY')
plt.hist(py,bins,rwidth=0.8,histtype='bar', color='m',orientation='vertical',label='ML')
plt.title('Python/Machine Learning Learner')
plt.xlabel('Student Age')
plt.ylabel('No. of Students (AGE)')
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize=(16,9))
bins = [10,15,20,25,30,35,40,45]
style.use('ggplot')
plt.hist([ml,py],bins,rwidth=0.8,histtype='bar', color=['r','y'],orientation='vertical',label=['PY','ML'])
#plt.hist(py,bins,rwidth=0.8,histtype='bar', color='m',orientation='vertical',label=['ML','PY'])
plt.title('Python/Machine Learning Learner')
plt.xlabel('Student Age')
plt.ylabel('No. of Students (AGE)')
plt.legend(loc=2)
plt.grid(color='k', linestyle='--')
plt.show()
```

# Bar Chart

```python
#We are creating a dummy dataset
classes = ['Python','R','Artificial Intelligence','Machine Learning']
pres_std_cls_1 = [10,30,45,34]
pres_std_cls_2  =[34,10,34,7]
pres_std_cls_3 = [12,10,29,2]
```

```python
plt.figure(figsize=(12,5))
plt.bar(classes,pres_std_cls_1,width=0.2,edgecolor='w',color='r',linewidth=5,alpha=0.6,linestyle='--',label='Analysis')
plt.title('Analysis')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.barh(classes,pres_std_cls_1,edgecolor='w',color='r',linewidth=5,alpha=0.6,linestyle='--',label='Analysis')
plt.title('Analysis')
plt.show()
```

```python
plt.figure(figsize=(16,9))
class_index = np.arange(len(classes))
width = 0.2
plt.bar(class_index,pres_std_cls_1,width,label='Class1')
plt.bar(class_index+width,pres_std_cls_2,width,label='Class2',color='b')
plt.xticks(class_index+width,classes)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(11,5))
class_index = np.arange(len(classes))
width = 0.2
plt.bar(class_index,pres_std_cls_1,width,label='Class1')
plt.bar(class_index+width,pres_std_cls_2,width,label='Class2',color='b')
plt.bar(class_index+width+width,pres_std_cls_3,width,label='Class3',color='y')
plt.xticks(class_index+width,classes,fontsize=9,rotation=30)
plt.legend()
plt.show()
```

```python
#We are creating a dummy dataset
classes = ['Python','R','Artificial Intelligence','Machine Learning']
pres_std_cls_1 = [10,30,45,34]
pres_std_cls_2  =[34,10,34,7]
pres_std_cls_3 = [12,10,29,2]
```

```python
plt.figure(figsize=(10,6))
class_index  = np.arange(len(classes))
width = 0.2
plt.bar(class_index,pres_std_cls_1,width, label='Class 1')
plt.bar(class_index+width,pres_std_cls_2,width, label='Class 2')
plt.bar(class_index+width+width,pres_std_cls_3,width, label='Class 3')
plt.title('Analysis b/w students',fontsize=15)
plt.xticks(class_index+width,classes,rotation=30)
plt.legend()
plt.show()
```
