```python
df1  =pd.DataFrame({'ID':[1,2,3,4],'Class':[9,10,11,12]})
df2 = pd.DataFrame({'ID':[1,2,3,5],'Name':['A','B','C','D']})
```

```python
pd.merge(df1,df2,on='ID')
```

```python
pd.merge(df1,df2,on='ID',how='left') #this takes by default inner we'll use some other like outer, left and right, left will take 1st Dataframe
```

```python
pd.merge(df1,df2,on='ID',how='right')
```

```python
pd.merge(df1,df2,on='ID',how='outer') # this will take both ID's
```

```python
pd.merge(df1,df2,on='ID',indicator=True,how='outer') # indicator will indicate the position in dataframe
```
